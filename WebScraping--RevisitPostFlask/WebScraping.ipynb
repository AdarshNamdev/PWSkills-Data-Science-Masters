```python
from bs4 import BeautifulSoup as bs
import requests
from urllib.request import urlopen
```

```python
flipkart_url = "https://www.flipkart.com"
flipkart_search_ulr = flipkart_url+"/search?q="

flipkart_search_ulr
```

```python
httpclient_obj = urlopen(flipkart_search_ulr+"tv")
httpclient_obj
```

```python
flipkart_page = httpclient_obj.read()
flipkart_page
```

```python
html_beautified = bs(flipkart_page, features= 'html.parser')
html_beautified
```

```python
html_box = html_beautified.find_all(name= "div", attrs= {"class": "_1AtVbE col-12-12"})
print(len(html_box))
html_box
```

```python
del html_box[0:3]

print(len(html_box))

html_box
```

```python
html_box[0].div.div.div.a['href']
```

```python
# appending the above url to the base flipkart page url

product_page = flipkart_url+html_box[0].div.div.div.a['href']
product_page
```

```python
# getting all the product pages inside a list

product_page_link = []
for i in range(0, len(html_box)):
    #print(i)
    try:
        page_urls = flipkart_url+html_box[i].div.div.div.a['href']
        product_page_link.append(page_urls)
    
    except AttributeError:
        print(f"AttributeError at index - {i}: href not present !")

    
product_page_link    
```
