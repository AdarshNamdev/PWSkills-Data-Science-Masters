#### Q 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Below are the five plots that can be created using seaborn library are:

`import seaborn as sb`

* histogram     `sb.histplot()`

Histogram plot is used to obtain the frequency/count of values under the predictor/column

* barplot       `sb.barplot()`

A common plot used to visualize the frequencies in categorical data is the bar plot, where we display the observed values as the heights of bars. All of the bars have the same width

* scatterplot   `sb.scatterplot()`

Scatterplots are used to identify the pattern of correlation between two variables in the dataset. That means, it helps us identify how one variable changes according to another variable.

* boxplot       `sb.boxplot()`

A box plot is a statistical graphic that display the distribution of a dataset through 5 key values:
    * the minimum
    * first quartile
    * median
    * third quartile
    * Maximum
```
    Q1-1.5IQR   Q1   median  Q3   Q3+1.5IQR
                  |-----:-----|
  o      |--------|     :     |--------|    o  o
                  |-----:-----|
flier           <----------->            fliers
                       IQR
```

The features of a boz plot are:


The box, showing the range between the first and third quartile ( the Inter-Quartile Range (IQR) )

A Line, inside the box representing the Median

WHiskers, extending from the box to the minimum and maximum values (excluding outliers)

Potential outliers, shown as individual points beyond the whiskers

A Box plot provide a concise visual summary of the central tendancy, spread and skewness of a dataset.

* pairplot `sb.pairplot()`

pairplot are used when we want to see the relationship between every single predictor, scatter plots are plotted in one single figure and the plots inthe diagonal shows the distributioin of the predictor

#### Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

````python
import seaborn as sb
import matplotlib.pyplot as plt

fmri = sb.load_dataset('fmri')

fmri.head()
````

````python
plt.figure(figsize= (8,5))

sb.lineplot(x = fmri['timepoint'], y = fmri['signal'], hue = fmri['region'], style = fmri['event'])
plt.show()
````

#### Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

````python
titanic = sb.load_dataset('titanic')
titanic.head()
````

````python
plt.subplot(121)
sb.boxplot(x = titanic['pclass'], y = titanic['age'])

plt.subplot(122)
sb.boxplot(x = titanic['pclass'], y = titanic['fare'])

plt.tight_layout()
plt.show()
````

#### Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

````python
gem = sb.load_dataset('diamonds')
gem.head()
````

````python
plt.figure(figsize= (20, 8))
sb.histplot(data = gem, x= 'price', hue = 'cut')

plt.show()
````

#### Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

````python
iris = sb.load_dataset('iris')
iris.head()
````

#### Q6: Use the "flights" dataset from seaborn to plot a heatmap.

````python
flights = sb.load_dataset("flights")
flights.head()
````

````python
import calendar

month_map = {}
for i in range(1, 13):
    month_map[calendar.month_name[i][0:3]] = i
    
print(month_map)

month_num = flights.month.map(month_map)

flights['month_num'] = month_num

flights = flights[['year', 'month_num', 'passengers']]
flights
````

````python

sb.heatmap(data = flights, cmap = 'Blues_r')
plt.show()
````
