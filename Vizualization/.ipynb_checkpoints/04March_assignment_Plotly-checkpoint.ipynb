#### Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import plotly.express as exp
import seaborn as sb
import numpy as np
```

```python
titanic = sb.load_dataset('titanic')
titanic.head()
```

```python
scat = exp.scatter(data_frame = titanic,
                   x = titanic.age,
                   y = titanic.fare,
                   color = titanic.who,
                   title = "Titanic Scatter Plot",
                                     
                  )

scat
```

#### Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
tips = sb.load_dataset("tips")
tips.head()
```

```python
hist = px.histogram(data_frame = tips, x = 'sex', y = 'total_bill', pattern_shape = 'smoker', color = 'day')
hist
```

#### Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

```python
iris = sb.load_dataset('iris')
iris.head()
```

```python
px.scatter_matrix(data_frame = iris,
                  color = 'species',
                  dimensions = ["sepal_length", "sepal_width", "petal_length", "petal_width"],
                 width = 1000,
                 height = 1000)
```

#### Q5. What is Distplot? Using Plotly express, plot a distplot.

Distplot is a plot that lets know the how the data is distributed, is it normally distrubuted or have some other type of distribution.

```python
import plotly.graph_objects as go
import plotly.figure_factory as ff
```

```python
fig = ff.create_distplot(hist_data = [iris.sepal_length.values.tolist(), iris.sepal_width.values.tolist()],
                        group_labels = ['sepal length', 'sepal_width'],
                        colors = ['#A80874', '#2B9720'],
                        show_rug= False
                        )
fig.layout.update(title = 'iris sepal distribution')
fig
```
