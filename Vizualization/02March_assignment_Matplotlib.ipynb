#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
```
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Note: Also add title, xlabel, and ylabel to the plot.
```

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

ax = plt.axes()
ax.set_facecolor("#C6E0FF")
ax.scatter(x = x, y = y, marker = 'x')
ax.set_title("Scatter plot x Vs. y")
ax.set_xlabel("x values ----->")
ax.set_ylabel("y values ----->")

fig = plt.gcf()
fig.set_size_inches(10, 5)

plt.show()
````

#### Q3: Why is the `subplot()` function used? Draw four line plots using the `subplot()` function.

Use the following data:

```
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
```

Subplot function can be used to have different/multiple plot on the same figure. Subplot function is useful is comparing plots side by side plotted with same or different dataset. User can plot any type of plot in different subplots at the same time.

````python
import numpy as np
#For line 1: 
ax1 = plt.subplot(221)
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
ax1.plot(x1, y1)
ax1.set_title("Subplot - 1")



#For line 2:
ax2 = plt.subplot(222)
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
ax2.plot(x2, y2)
ax2.set_title("Subplot - 2")


#For line 3:
ax3 = plt.subplot(223)
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
ax3.plot(x3, y3)
ax3.set_title("Subplot - 3")


#For line 4:
ax4 = plt.subplot(224)
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
ax4.plot(x4, y4)
ax4.set_title("Subplot - 4")


plt.tight_layout()
plt.show()
````

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
```
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

A common plot used to visualize the frequencies in categorical data is the bar plot, where we display the observed values as the heights of bars. There are 2 different way of representation of bar charts, vertiacally alligned and horizontally alligned.

````python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


plt.subplot(211)
plt.bar(x = company, height = profit, color = '#574AE2')
plt.grid('dark')
plt.xlabel("Organizations")
plt.ylabel("Profit")
plt.title("BAR CHART")

plt.subplot(212)
plt.barh(y = company, width = profit, color = '#EA1744')
plt.grid('dark')
plt.ylabel("Organizations")
plt.xlabel("Profit")
plt.title("HORIZONTAL BAR CHART")



plt.tight_layout(h_pad = 0.5, w_pad = 0.9)
plt.show()
````

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
```
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A box plot is a statistical graphic that display the distribution of a dataset through 5 key values:
* the minimum
* first quartile
* median
* third quartile
* Maximum
```
    Q1-1.5IQR   Q1   median  Q3   Q3+1.5IQR
                  |-----:-----|
  o      |--------|     :     |--------|    o  o
                  |-----:-----|
flier           <----------->            fliers
                       IQR
```

The features of a boz plot are:


The box, showing the range between the first and third quartile ( the Inter-Quartile Range (IQR) )

A Line, inside the box representing the Median

WHiskers, extending from the box to the minimum and maximum values (excluding outliers)

Potential outliers, shown as individual points beyond the whiskers

A Box plot provide a concise visual summary of the central tendancy, spread and skewness of a dataset.

````python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.subplot(121)
plt.boxplot(box1)
plt.subplot(122)
plt.boxplot(box2)

plt.show()
````
