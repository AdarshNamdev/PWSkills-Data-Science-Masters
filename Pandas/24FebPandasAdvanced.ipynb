### Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
d = {'A':['Apirin', 'Minoxydile', 'cph4', 'thymine', 'betadine'],
     'B' : ['ranbaxy', 'cipla', 'glaxosmithkline', 'AstraZeneca', 'Pfizer'],
     'C': [1954, 1999, 1977, 1988, 1847]
    }



df = pd.DataFrame(d)
df.head()
```

```python
def reindex_dataframe(dframe):
    new_index = pd.Series(range(1, 2*len(dframe), 2))
    return dframe.set_index(new_index)

reindex_dataframe(df)
```
