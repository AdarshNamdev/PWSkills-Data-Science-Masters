### Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
d = {'A':['Apirin', 'Minoxydile', 'cph4', 'thymine', 'betadine'],
     'B' : ['ranbaxy', 'cipla', 'glaxosmithkline', 'AstraZeneca', 'Pfizer'],
     'C': [1954, 1999, 1977, 1988, 1847]
    }



df = pd.DataFrame(d)
df.head()
```

```python
def reindex_dataframe(dframe):
    new_index = pd.Series(range(1, 2*len(dframe), 2))
    return dframe.set_index(new_index)

reindex_dataframe(df)
```

```python
df.set_index(pd.Series(range(1,2*len(df), 2)))
```

### Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The function should print the sum to the console. 

### For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should calculate and print the sum of the first three values, which is 60.

```python
q3 = pd.DataFrame({'values' : [10, 20, 30, 40, 50]})
q3
```

```python
def sum_first_three(dframe):
    summation = sum(dframe.values.flatten()[0:3])
    
    return summation

sum_first_three(q3)
```

```python
# instead of writing a function we can achieve the same by just using the below construct

summation_2 = sum(q3.values.flatten()[0:3])
print(summation_2)


# alterntively we can also directly do :

q3['values'].head(3).sum()
```

### Q5. How are `DataFrame.size` and `DataFrame.shape` different?

`DataFrame.size` and `DataFrame.shape` both are the attributes of the pandas `DataFrame` class.

`DataFrame.size` return an integer which tells the total number of elements (rows * columns) present in the dataframe. Where as `DataFrame.shape` return a two element tuple whose 0th index element repesents the number of rows and 1st index element represents the total number of columns present in the dataframe.

### Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

```python
np.datetime64('2012-02-14'), np.datetime64('2012-02-14')+1
```

```python
datasample = {'date': [np.datetime64('2012-02-14') + i for i in range(0,20)],
             'sales':np.random.rand(20,)}

df10 = pd.DataFrame(datasample)
df10.head(10)
```

```python
df10['MovingAverage'] = df10['sales'].rolling(window = 7, min_periods= 0).mean()

df10.head(20)
```

```python
df10[df10['MovingAverage'] > 0.55]
```
