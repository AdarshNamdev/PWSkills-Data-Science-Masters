```python
perc = float(input("Enter percentage: "))

if perc < 100:
    if perc < 60:
        print("Grade: D")
    
    if 60 <= perc <= 80:
        print("Grade: C")
    
    if 80 < perc <= 90:
        print("Grade: B")
        
    if perc > 90:
        print("Grade: A")

else:
    print("Incorrect Input!")
```

```python
given_number = input("Enter a number greater than 10: ")

if given_number.isnumeric():
    if int(given_number) <= 10:
        print("you have to enter a number which is 'greater' than 10")
    
    else:
        number = int(given_number)
        """
        below loop will reduce the number by 1 and count how many times the updated number
        will get divided by 3 before its <= 10
        """
        division_counter = 0
        while number > 10:
            if number%3 == 0:
                division_counter += 1
            else:
                pass
            number = number - 1
        
        print(f"times a given number can be divided by 3 before it is less than or equal to 10 is: {division_counter}")
    
else:
    print("Invalid Input. Enter a whole Number!")
```

While loop in python can be used when we want to have our loop keep running based on a condition until its True.

```python
n = 10

for i in range(0, n):
    for j in range(0, i+1):
        print("* ", end = '')
    print("\r")
```

```python
# we are printing an equilateral triangle

side_len = int(input("Enter the side length of your equilateral triangle: "))

#spaces = side_len / 2

padding = " "
spaces = side_len
i = 1
while i <= side_len:
    print(" "*spaces, end = '')
    print(("*"+padding)*i)
    spaces = spaces - 1
    i = i+1
```
