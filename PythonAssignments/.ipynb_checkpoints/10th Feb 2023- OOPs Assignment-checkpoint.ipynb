### Q1. Which function is used to open a file? What are the different modes of opening a file? Explain each mode of file opening.

```open()``` is the function that is used to open a file.

Different modes of opening a file are:

* __"r"__   :: This mode is the default mode and used for opening a file in read-only mode
* __"w"__   :: this mode is used for writing the file that is opened. file content is truncated everytime file is re-opened and written using this mode
* __"r+"__  :: this is mode is used when a programmer wants to read as well write inside the file simultaneously without truncating
* __"w+"__  :: with this mode a file gets truncated first and data can be written in it. the file can also be read.
* __"rb"__  :: reading the file in binary format
* __"wb"__  :: openig the file to write binary format data in it.
* __"a"__   :: opening the file in 'append' mode. with this mode the cursor is palced at the end of file to put more information in it.
* __"wb+"__ :: openig the file to write and read binary format data in it.
* __"a+"__  :: openig the file for both appending and reading data

### Q2. Why close() function is used? Why is it important to close a file?

`close()` is used to close a file once the data is read/written in it. If the file is not closed it will remain opened and we wont be able to perform operations on it.

for example:
Consider the below code where we are opening a file in 'append' mode and written some content in the file.
```
f = open(r"D:\PWSkills_DS_Masters\pwskills.txt", "a+")
f.writelines("appending some more content in this file!")
```

now if the try to __delete__ this file from the directory... we wont be able to do so, as its already opened by the python code above



### Q3. Write a python program to create a text file. Write _'I want to become a Data Scientist'_ in that file. Then close the file. Open this file and read the content of the file.

````python
with open(r"D:\PWSkills_DS_Masters\pwskills.txt", "w+") as fh1:
    fh1.writelines("""I want ot become a Data Scientist""")
    fh1.seek(0)
    print(fh1.read())
````

### Q4. Explain the following with python code: `read()`, `readline()` and `readlines()`.

````python
with open(r"D:\PWSkills_DS_Masters\pwskills.txt", "r") as fh2:
    read_method = fh2.read()
    fh2.seek(0)
    readline_method = fh2.readline()
    readline_method2 = fh2.readline()
    fh2.seek(0)
    readlines_method = fh2.readlines()
    
    print(f"read(): \n{read_method}")
    print(f"\nreadline(): \n{readline_method}")
    print(f"read next line: \n{readline_method2}")
    print(f"\nreadlines(): \n{readlines_method}")
````

__explanation__:

`read()` method is used to read the entire content of the file at once!<br>
`readline()` method reads one line at a time from the file and the cursor then is positioned at the end of the line.<br>
`readlines()` with this method the content is readline by line and stored in the list and this list of lines are returned

### Q5. Explain why `with` statement is used with `open()`. What is the advantage of using `with` statement and `open()` together?

`with` is a context manager that works with `open()` function that open a file. `with` automatically closes the file after the operations on the file is no longer needed, thats the benefit of using `with`

````python
with open(r"D:\PWSkills_DS_Masters\pwskills.txt", "r") as fh2:
    a= fh2.read()
    print(a)
````

### Q6. Explain the `write()` and `writelines()` functions. Give a suitable example.

`write()` method is used to write one line at a time inside the file. once a line is written the cursor remains at the end of line and if again `write()` method is called, python writes the new content from there.

`writelines` with this method we can write multiple lines at once inside a file.

````python
with open(r"D:\PWSkills_DS_Masters\pwskills2.txt", "w+") as infile:
    position = []
    for num in range(0,3):
        pos = infile.write(f"{num+1}.) this is line #{num+1}\n")
        position.append(pos)
    infile.seek(0)
    file_content = infile.read()
    print(file_content)
    
    infile.writelines("""\nUsed writelines() method to added multiple lines
    at once inside the file. File handling is interesting, it seems!""")
    
    infile.seek(sum(position)+1)  # to put the cursor at the position where first content ends.
    new_content = infile.read()
    print(new_content)
````
