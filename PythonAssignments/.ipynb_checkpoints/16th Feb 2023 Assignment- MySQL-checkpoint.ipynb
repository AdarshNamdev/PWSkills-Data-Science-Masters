### Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.

In SQL, DDL stands for Deta Definition Language. These are the SQL statements which are used for determining the structure of the database. These commands are not usually executed/ran by the normal user and only specialized role like Database Administrator, can execute them.

__CREATE__: command is used for creating databases and their underlying objects like Tables, Views, Stored Procedures etc.
```
CREATE TABLE PHONE
(
productionID INTEGER,
brand varchar2(50),
model varchar2(50),
date_of_mfg DATE,
)
```

__DROP__: commnad is used for deleting the objects from the database.

`DROP TABLE PHONE`

__ALTER__: at some point in time if there is modification required in the structure of database and/or its objects then `ALTER` command is used. Example `ALTER` command is used for adding a column, renaming a column, dropping a column or renamng a table.

=======================================================================================<br>
_EXAMPLE #1 :: Adding columns_

```
ALTER TABLE PHONE
ADD network_supported varchar2(10) not null,
ADD operating_system varchar2(20) not null,
ADD country_of_origin char(50),
ADD color char(50)
ADD phone_condition char(50);
```

_EXAMPLE #2 :: Dropping columns_

```
ALTER TABLE PHONE
DROP COLUMN country_of_origin
```

_EXAMPLE #3 :: modifying columns (i.e changing the datatype and adding/removing the null -notnull constraint)_

```
ALTER TABLE PHONE
MODIFY country_of_origin varchar(50) not null,
MODIFY color char(20);

```

_EXAMPLE #4 :: Renaming columns_

```
ALTER TABLE PHONE
CHANGE COLUMN operating_system OS varchar2(20);

```

_EXAMPLE #4 :: Renaming table_

```
ALTER TABLE PHONE
RENAME TO SMARTPHONES;
```

=======================================================================================<br>


__TRUNCATE__: command is used to remove/purge the data present in the database tables. `TRUNCATE` also removes the space allocated for storing the table records.

`TRUNCATE TABLE SMARTPHONES`

### Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.

__DML__ stands for Data Manipulation Language, a subset of SQL used to query, insert, update, and delete data in a database. DML statements are the most commonly used SQL statements, as they allow you to perform all of the basic operations that you need to do with data in a database.

The four main DML statements are:

`INSERT`: This statement is used to insert new data into a database table.<br>
`UPDATE`: This statement is used to update existing data in a database table.<br>
`DELETE`: This statement is used to delete data from a database table.<br>

a.) `INSERT` statement
```
INSERT INTO customers
(first_name, last_name, email, address) 
VALUES ('Adarsh', 'Namdev', 'adarsh.namdev@somedomain.com', "Street 11, Berlin");
```

b.) `UPDATE` statement
```
UPDATE customers 
SET address = "Street 11, Frankfurt"
WHERE id = 1;
```

c.) `DELETE` statement
```
DELETE FROM customers 
WHERE id = 2;
```

### Q4. What is DQL? Explain SELECT with an example.

__DQL__ stands for Data Query Language. It is a subset of the SQL language that is used to retrieve data from a database. DQL statements are used to select and filter data from one or more database tables.

The `SELECT` statement is the most important DQL statement. It is used to specify the columns and rows that you want to retrieve from a database table. The `SELECT` statement can also be used to filter the data that is retrieved by using the `WHERE` clause.

```
SELECT title, name, email
FROM customers
WHERE country = 'WESTEROS';
```

### Q5. Explain Primary Key and Foreign Key.

Primary Key and Foreign Key are two type of constraints in databases. 
A primary key is a column in the table that is unique identifier w.r.t every row whose value can never be duplicate. Primary Key column can never hold a NULL value i.e. at the time of creating a table we have to apply `NOT NULL` constaint against this column.

On the other hand, Foreign Key are the columns which are used to establish a relationship/connection between two or more tables by referencing the PRIMARY KEY of the other table. At the time of creating a table we designate a column to be the Foreign Key that will help connect with other tables using `JOINS`. 

### Q6. Write a python code to connect MySQL to python. Explain the cursor() and execute() method.

`cursor()` is method that is called on the mysql connection object that allows us run a SQL query.

`execute()` is method that provided by the cursor object inorder to run a SQL query which is passed as the argument to it.

````python
import mysql.connector as con

mydb = con.connect(
    host = "localhost",
    user = "abc",
    password = "password"
)

print(mydb)
cur = mydb.cursor()
cur.execute("""
            CREATE Database IF NOT EXISTS sample_db
            """)

mydb.commit()
mydb.close()
````
